```python
radial=np.sqrt(xx**2 + yy**2)
angle=np.arctan2(yy,xx)
radial, angle
```

```python
plt.pcolormesh(xx,yy, radial)
```

```python
plt.pcolormesh(xx,yy, angle)
```

```python
f = (np.cos(4*radial))**2 + (np.sin(4*angle))**2

plt.pcolormesh(xx,yy, f)
```

```python
f.mean()

f_xmean = f.mean(axis=1)
plt.plot(f_xmean, y);
y
f_xmean
```

```python
f_ymean = f.mean(axis=0)
plt.plot(x, f_ymean);
f_ymean
x
```

#### 2.6 Account For Missing Data

The profiles contain many missing values. These are indicated by the special “Not a Number” value, or np.nan.

When you take the mean or standard deviation of data with NaNs in it, the entire result becomes NaN. Instead, if you use the special functions np.nanmean and np.nanstd, you tell NumPy to ignore the NaNs.

Recalculate the means and standard deviations as in the previous sections using these functions and plot the results.

```python
newmeanT = np.nanmean(T, axis=1)
newstdT = np.nanstd(T, axis=1)
plt.plot(newmeanT, levels)
plt.errorbar(newmeanT, levels, xerr = newstdT)
plt.gca().invert_yaxis()
plt.grid (True, c = 'black')
```

```python
newmeanS = np.nanmean(S, axis=1)
newstdS = np.nanstd(S, axis=1)
plt.plot(newmeanS, levels)
plt.errorbar(newmeanS, levels, xerr = newstdS)
plt.gca().invert_yaxis()
plt.grid (True, c = 'black')
```

```python
newmeanP = np.nanmean(P, axis=1)
newstdP = np.nanstd(P, axis=1)
plt.plot(newmeanP, levels)
plt.errorbar(newmeanP, levels, xerr = newstdP)
plt.gca().invert_yaxis()
plt.grid (True, c = 'black')
```
